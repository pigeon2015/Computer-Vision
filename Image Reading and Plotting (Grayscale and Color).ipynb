```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image in color
image_color = cv2.imread("D:/photoForComputerVision/rabit.jpg", cv2.IMREAD_COLOR)
```

```python
# Convert BGR image to RGB
image_color_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Read the image in grayscale
image_grayscale = cv2.imread("D:/photoForComputerVision/rabit.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
# Display the images using matplotlib
plt.figure(figsize=(10, 5))
```

```python
# Display color image
plt.subplot(1, 2, 1)
plt.imshow(image_color_rgb)
plt.title('Color Image')
plt.axis('off')
```

```python
# Display grayscale image
plt.subplot(1, 2, 2)
plt.imshow(image_grayscale, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```
